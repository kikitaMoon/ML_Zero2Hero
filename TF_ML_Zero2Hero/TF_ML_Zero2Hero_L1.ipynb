to find the relationship between X and Y:
- X = -1, 0, 1, 2, 3, 4
- Y = -3, -1,1, 3, 5, 7

Y = 2X - 1


```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
# Define the model itself,  it contains single layer (keras.layers.Dense) with single neuron (units=1) in it.
# input_shape is for X

model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])]) 
```

```python
# optimizer to generate a guess,  and loss function to know how good or bad the guess is .
# then the optimizer will improve based on the result of loss function

model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# input data is defined as array 

xs = np.array([-1.0,0.0, 1.0,2.0,3.0,4.0], dtype=float) 
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float) 
```

```python
# model will run 500 times to train the model,  that is, fit X to Y and try 500 times.

model.fit(xs, ys, epochs=500)
```

```python
# when X= 10, then Y = ?  it is expected to be close to 19

print(model.predict([10.0] ) ) 
```
